```python
# Какой части пассажиров удалось выжить? Посчитайте долю выживших пассажиров. 
# Ответ приведите в процентах (число в интервале от 0 до 100, знак процента не нужен), округлив до двух знаков.
```

```python
Survived = (data['Survived']==1).sum()
All = (data['Survived']).sum() + (data['Survived']==0).sum()
```

```python
def toFixed(numObj, digits=0):
    return f"{numObj:.{digits}f}"
```

```python
percentage = Survived / All*100
print(toFixed(percentage, 2))
```

```python
# Какую долю пассажиры первого класса составляли среди всех пассажиров? 
# Ответ приведите в процентах (число в интервале от 0 до 100, знак процента не нужен), 
# округлив до двух знаков.
```

```python
First_class = (data['Pclass']==1).sum()
All_passengers = (data['Pclass']!=1).sum() + First_class
```

```python
First_class
```

```python
All_passengers
```

```python
percentage = First_class / All_passengers*100
print(toFixed(percentage, 2))
```

```python
median, mean = data['Age'].median(), data['Age'].mean()
print(median, toFixed(mean, 2))
```
