```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
data = pd.read_csv('close_prices.csv')
```

```python
data.head()
```

```python
X = data.drop('date', axis=1)
```

```python
# На загруженных данных обучите преобразование PCA с числом компоненты равным 10. 
# Скольких компонент хватит, чтобы объяснить 90% дисперсии?
```

```python
pca = PCA(n_components=10)
```

```python
pca.fit(X)
```

```python
s = pca.explained_variance_ratio_
```

```python
k = pca.components_[0]
```

```python
k = np.array(k)
```

```python
k
```

```python
maxim = k[0]
for ind, value in enumerate(k):
    if value>maxim:
        maxim = value
print(maxim)
print(k.index(maxim))
```
