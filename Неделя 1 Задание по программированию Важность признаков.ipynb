```python
# Загрузите выборку из файла titanic.csv с помощью пакета Pandas.
```

```python
import pandas as pd
data = pd.read_csv('titanic.csv', sep=',')
import numpy as np
```

```python
data.head(10)
```

```python
# Оставьте в выборке четыре признака: класс пассажира (Pclass), цену билета (Fare), 
# возраст пассажира (Age) и его пол (Sex).
```

```python
data_new = data[['Pclass', 'Fare', 'Age', 'Sex']]
```

```python
data_new.head()
```

```python
# Обратите внимание, что признак Sex имеет строковые значения.
```

```python
def string_gender(x):
    return 1 if x == 'female' else 2
```

```python
data['Female']=data['Sex'].apply(string_gender)
```

```python
data_new = data[['Pclass', 'Fare', 'Age','Female', 'Survived']]
```

```python
data_new.head(50)
```

```python
data_new.shape
```

```python
# В данных есть пропущенные значения — например, для некоторых пассажиров неизвестен их возраст. 
# Такие записи при чтении их в pandas принимают значение nan. Найдите все объекты, у которых есть 
# пропущенные признаки, и удалите их из выборки.
```

```python
data_new_1 = data_new.dropna()
```

```python
data_new_1.shape
```

```python
data_new_1.head()
```

```python
# Выделите целевую переменную — она записана в столбце Survived
```

```python
y = data_new_1['Survived']
```

```python
del data_new_1['Survived']
```

```python
x = data_new_1
x.head()
```

```python
# Обучите решающее дерево с параметром random_state=241 и остальными параметрами по умолчанию 
# (речь идет о параметрах конструктора DecisionTreeСlassifier).
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=241)
clf.fit(x, y)
```

```python
importances = clf.feature_importances_
```
